## Q5. How do you choose which type of Naive Bayes classifier to use for any given problem?

- If in the given problem the dataset follows Bernoulli distribution i.e output can only be true or false/Success or Failure,then we will use Bernoulli Naive Bayes classifier.
- If the problem has text in Input Feature then we will use Multinomial Naive Bayes in which we will first use NLP to convert text into numerical data to solve the problem.
- If in the given problem dataset follows Gaussian Distribution then we will use Gaussian Naive Bayes Classifier
